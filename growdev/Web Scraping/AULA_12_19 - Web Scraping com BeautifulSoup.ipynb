# Realizando requests e obtendo respostas HTML

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.quintoandar.com.br/alugar/imovel/curitiba-pr-brasil?flexible=true&referrer=profilingv2'
response = requests.get(url)
```

```python
response
```

# Navegação básica e extração de dados

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
section = soup.select_one(".SideMenuHouseList_customSection__XJ5QW")

if section:
    print(section.prettify()) 
else:
    print("Elemento não encontrado.")
```

# Localizando elementos com find() e find_all()

```python
imoveis = section.find_all('div', {'data-testid': 'house-card-container-rent'})

print(imoveis)
```

```python
for imovel in imoveis:
    
    print(f'imovel: {imovel}')
```

# Extração de tabelas, listas e links

```python
for imovel in imoveis:
    link = imovel.find('a', {'class': 'StyledLink_styledLink__P_6FN'})['href']
    url_completa = "https://www.quintoandar.com.br" + link 
    
    print(f'url: {url_completa}')
```

# Manipulação de textos e tratamento de dados

```python
import re
```

```python
imoveis = []

for imovel in soup.find_all('div', {'data-testid': 'house-card-container-rent'}):
    
    link = imovel.find('a', {'class': 'StyledLink_styledLink__P_6FN'})['href']
    ds_url = "https://www.quintoandar.com.br" + link 
    ds_titulo = imovel.find('a', {'class': 'StyledLink_styledLink__P_6FN'})['title']
    
    aria_label = imovel.find('div', class_='Cozy__FindHouseCard-Container')
    
    if aria_label and 'aria-label' in aria_label.attrs:
        info_text = aria_label['aria-label']
        
        ds_endereco = info_text.split("metros quadrados")[0].split(".")[-2].strip()
        
        qt_area = info_text.split("metros quadrados")[0].split(".")[-1].strip()
        
        match_quartos = re.search(r'(\d+)\s+quarto[s]?', info_text)
        qt_quartos = match_quartos.group(1) if match_quartos else "0"
        
        match_garagens = re.search(r'(\d+)\s+vaga[s]? de garagem', info_text)
        qt_garagens = match_garagens.group(1) if match_garagens else "0"
        
        vl_aluguel = info_text.split("aluguel")[0].split("R$")[-1].strip()
        
        vl_total = info_text.split("total")[0].split("R$")[-1].strip()
        
        imoveis.append({
            'link': ds_url,
            'título': ds_titulo,
            'endereco': ds_endereco,
            'área': qt_area,
            'quartos': qt_quartos,
            'garagens': qt_garagens,
            'aluguel': vl_aluguel,
            'total': vl_total
        })

print(imoveis)
```

# Salvando os dados em CSV e JSON

```python
import csv
import json
```

```python
with open('imoveis.csv', mode='w', newline='', encoding='utf-8') as csv_file:
    fieldnames = imoveis[0].keys() 
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(imoveis)
```

```python
with open('imoveis.json', mode='w', encoding='utf-8') as json_file:
    json.dump(imoveis, json_file, ensure_ascii=False, indent=4)
```
